# **Nutrient analysis**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
dataset = pd.read_csv('https://query.data.world/s/fkumtivouh5oejiblmop6dip24hlgq')
dataset
```

```
#import preprocessing from sklearn
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```
df = dataset
df['brand'] = le.fit_transform(dataset.brand)
df['brand'].unique()
df
```

# **Groupby brand A to check brand A**

```
gpa = dataset.groupby("brand")
gpa.get_group("A")
```

```
#drop id as it is not needed
dataset.drop(["id"],axis = 1,inplace = True)
```

```
dataset_agg = dataset.groupby("brand").sum()
dataset_agg
```

```
#without brand name for pca
dataset_no_id = dataset.drop(["brand"],axis = 1)
scaler_input_data = dataset_no_id
scaler_input_data
```

# **PCA**

```
#standard scaler
from sklearn.preprocessing import StandardScaler
```

```
scaler = StandardScaler()
scaler.fit(scaler_input_data)
pca_in = scaler.transform(scaler_input_data)
```

```
pca_in.shape
```

```
#pca
from sklearn.decomposition import PCA
```

```
#components
pca = PCA(n_components = 2)
pca.fit(pca_in)
```

```
pca_out = pca.transform(pca_in)
pca_out
```
